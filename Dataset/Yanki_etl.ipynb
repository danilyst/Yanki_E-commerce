```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
Yanki_df = pd.read_csv(r"C:\Users\Hp\Desktop\Yanki_E-commerce_Case_Study\Dataset\Raw_Dataset\yanki_ecommerce.csv")
```

```python
Yanki_df.info()
```

### Data Cleaning and Transformation


```python
#Dropping Rows that contain missing values
Yanki_df = Yanki_df.dropna()
Yanki_df.info()
```

```python
# Changing the data type of 'Order_Date'  to Datetime
"""" 
The 'dayfirst=True' parameter is used to specify that the day comes before the month in the date format.This is important for correctly parsing 
dates in formats like 'DD/MM/YYYY'.
"""
Yanki_df['Order_Date'] = pd.to_datetime(Yanki_df['Order_Date'], dayfirst=True)
```

### Normalization of Yanki_Df

```python
# Customer Table
customers_df = Yanki_df[['Customer_ID', 'Customer_Name', 'Email', 'Phone_Number']].copy().drop_duplicates().reset_index(drop=True)
```

```python
#Product Table
products_df = Yanki_df[['Product_ID', 'Product_Name', 'Brand', 'Category', 'Price']].copy().drop_duplicates().reset_index(drop=True)
```

```python
#Shipping_Address Table
shipping_address_df = Yanki_df[['Shipping_Address', 'City', 'State', 'Postal_Code', 'Country']].copy().drop_duplicates().reset_index(drop=True)
shipping_address_df.index.name = 'Shipping_Address_ID'
shipping_address_df = shipping_address_df.reset_index()
```

```python
#Orders Table
orders_df = Yanki_df[['Order_ID', 'Customer_ID', 'Product_ID', 'Quantity', 'Total_Price', 'Order_Date']].copy().drop_duplicates().reset_index(drop= True)
```

```python
#Payment_method table
payment_method_df = Yanki_df[['Order_ID', 'Payment_Method', 'Transaction_Status']].copy().drop_duplicates().reset_index(drop=True)   
```

```python
#Saving the tables as CSV files

customers_df.to_csv(r'C:\Users\Hp\Desktop\Yanki_E-commerce_Case_Study\Dataset\Final_Dataset\customers.csv', index = False)
products_df.to_csv(r'C:\Users\Hp\Desktop\Yanki_E-commerce_Case_Study\Dataset\Final_Dataset\products.csv', index = False)
shipping_address_df.to_csv(r'C:\Users\Hp\Desktop\Yanki_E-commerce_Case_Study\Dataset\Final_Dataset\shipping_address.csv', index = False)
orders_df.to_csv(r'C:\Users\Hp\Desktop\Yanki_E-commerce_Case_Study\Dataset\Final_Dataset\orders.csv', index = False)
payment_method_df.to_csv(r'C:\Users\Hp\Desktop\Yanki_E-commerce_Case_Study\Dataset\Final_Dataset\payment_method.csv', index = False)
```
